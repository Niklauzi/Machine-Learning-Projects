## Breast Cancer Wisconsin (Diagnostic)
https://archive.ics.uci.edu/dataset/17/breast+cancer+wisconsin+diagnostic

### **Problem Statement:**

Breast cancer is one of the leading causes of mortality among women worldwide. Early and accurate diagnosis of breast cancer is critical to improving patient outcomes and survival rates. This project aims to build a machine learning model to classify breast tumor types into malignant (cancerous) or benign (non-cancerous) based on features derived from clinical tests. 

The primary goal is to improve diagnostic accuracy and minimize false negatives (malignant cases incorrectly classified as benign) to ensure patients receive timely treatment. Additionally, balancing the dataset and optimizing model performance metrics like recall and precision are key objectives.

---

### **Hypotheses:**

#### **Null Hypothesis (H₀):**
The features derived from clinical tests (e.g., texture, perimeter, area, smoothness, etc.) do not provide enough information to accurately classify tumors as malignant or benign using machine learning models.

#### **Alternative Hypothesis (H₁):**
The features derived from clinical tests provide enough information to accurately classify tumors as malignant or benign using machine learning models, achieving a recall score above a predefined threshold (e.g., 90%) for the malignant class. 

---

### **Objectives:**
1. Develop a classification model to predict tumor types (malignant or benign) based on the given dataset.
2. Balance the dataset to improve model performance, particularly the recall score for malignant tumors.
3. Evaluate the model using appropriate metrics (confusion matrix, recall, precision, F1-score, etc.) to minimize false negatives.

### Data Preprocessing

```python
# No missing values was observed after carrying out EDA
```

```python
# Removing outliers, a number of outliers was observe in the dataset
```

```python
# Handling Outliers

def remove_outliers(df):
    for col in df.select_dtypes(include='number').columns:
        q1, q3 = df[col].quantile([0.10, 0.90])
        iqr = q3 - q1
        lower, upper = q1 - 1.5 * iqr, q3 + 1.5 * iqr
        df = df[(df[col] >= lower) & (df[col] <= upper)]
    return df
df_cleaned = remove_outliers(df)
df_cleaned
```

0.10 and 0.90 was used for q1 and q3 so as to be conservative and to detect only very extreme values, also the features were computed by a precision device meaning inputs should be mostly authentic. 

```python
# checking for missing value after handling outliers
df_cleaned.isna().sum()
```

### Feature Engineering

```python

# Separate the non-triplicate column (Diagnosis)
diagnosis = df_cleaned['Diagnosis']

# Group features by their base names (e.g., 'radius', 'texture')
feature_groups = {}
for col in df_cleaned.columns:
    if col != 'Diagnosis':  # Skip non-triplicate columns
        base_name = ''.join(filter(str.isalpha, col))  # Extract feature base name (e.g., 'radius')
        feature_groups.setdefault(base_name, []).append(col)

# Compute averages
averaged_features = {
    feature: df_cleaned[cols].mean(axis=1) for feature, cols in feature_groups.items()
}

# Create a new DataFrame with the averaged features and include the Diagnosis column
df_averaged = pd.DataFrame(averaged_features)
df_averaged['Diagnosis'] = diagnosis

df_averaged.head()
```

```python
df_averaged.info()
```

```python
averaged_features_columns = df_averaged.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = df_averaged[averaged_features_columns].corr()
```

```python
correlation_matrix.radius
```

```python
df = df_averaged.copy()
```

```python
"""We have to split the dataset to prevent data leakages when carrying out Encoding and Scaling"""

# Splitting the data into training and testing datasets
train_df, test_df = train_test_split(df, test_size=0.2, stratify=df['Diagnosis'], random_state=42)
```

```python
te = test_df.copy()
```

```python
# Initializing the LabelEncoder
le = LabelEncoder()


# Fit on the training set and transform both sets
train_df['Diagnosis'] = le.fit_transform(train_df['Diagnosis'])
test_df['Diagnosis'] = le.transform(test_df['Diagnosis'])
```

```python
train_df
```

```python
test_df
```

```python
train_df.Diagnosis.value_counts()
```

```python
# Separate features (X) and target (y) for training
X_train = train_df.drop(columns='Diagnosis')
y_train = train_df['Diagnosis']

# Separate features (X) and target (y) for testing
X_test = test_df.drop(columns='Diagnosis')
y_test = test_df['Diagnosis']
```

```python
# Standardizing
from sklearn.preprocessing import StandardScaler


scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)

X_test=scaler.transform(X_test)
```

```python
# Importing the model
from sklearn.linear_model import LogisticRegression

# Initializing model
logistic=LogisticRegression()

logistic.fit(X_train,y_train)
y_pred=logistic.predict(X_test)
print(y_pred)
```

```python
X_test.shape
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix

score=accuracy_score(y_test, y_pred)
print(score)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
test_df.Diagnosis.value_counts()
```

```python
print(le.classes_)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay


# Compute confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Display the confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Negative", "Positive"])
disp.plot(cmap="Blues")
```

### Hyperparameter Tuning And Cross Validation

### Grid SearchCV

```python
model=LogisticRegression()
penalty=['l1', 'l2', 'elasticnet']
c_values=[100,10,1.0,0.1,0.01]
solver=['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
```

```python
params=dict(penalty=penalty,C=c_values,solver=solver)
```

```python
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold()
```

```python
## GridSearchCV
from sklearn.model_selection import GridSearchCV

grid=GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv,n_jobs=-1)
```

```python
grid
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred1=grid.predict(X_test)
```

```python
score=accuracy_score(y_test, y_pred1)
print(score)
print(classification_report(y_test, y_pred1))
print(confusion_matrix(y_test, y_pred1))

cm = confusion_matrix(y_test, y_pred1)

# Display the confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Negative", "Positive"])
disp.plot(cmap="Blues")
```
